```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LinearRegression

# Create a pipeline with StandardScaler and LinearRegression
pipe = Pipeline([('scaler', StandardScaler()), ('regressor', LinearRegression())])

# Define the parameter grid for GridSearchCV
param_grid = {
    'regressor__copy_X': [True],
    'regressor__n_jobs': [-1, 1, 2, -2],
    'regressor__positive': [False],
    'regressor__fit_intercept': [True],
    # Add parameters for the scaler if needed, e.g., 'scaler__with_mean': [True, False]
}

# Create the GridSearchCV object
grid_model = GridSearchCV(pipe, param_grid=param_grid, n_jobs=-1, cv=5)

# Fit the model
grid_model.fit(xtrain, ytrain)
```

```python
grid_model.get_params()
```

```python
grid_model.best_params_
```

```python
grid_model.best_score_
```

```python
lasso = Lasso()
lasso.fit(xtrain,ytrain)
lasso.predict(xtest)
lasso.score(xtrain,ytrain)
lasso.score(xtest,ytest)
```

```python
lasso_grid_para={
    'alpha':[1.0,0.01,0.2,0.3,0.01,0.02],
    'fit_intercept':[True],
    'copy_X':[True],
    'max_iter':[1000,2000,5000,7000,10000,13000,15000],
    'positive':[False],
    'random_state':[0],
    'selection':['cyclic']
}
```
